```python
#----------------------------------------------------------------------
# Módulo 7 - Practico Nº2 
#
# Carlos Thieme B.
#
#----------------------------------------------------------------------

import pandas as pd 
import matplotlib.pyplot as plt 

def diagrama_ig(datos):
    fig, ax = plt.subplots()
    datos.plot(ax = ax)
    ax.set_ylim([0, max(datos.Ingresos.max(), datos.Gastos.max()) + 500])
    plt.title('Evolución ingresos y gastos')
    return ax

datos = {'Mes':['Ene', 'Feb', 'Mar', 'Abr', 'Mayo', 'Junio'], 'Ingresos':[53400, 25200, 14800, 65300, 75400, 66500], 'Gastos':[12300, 24450, 32000, 25200, 12300, 23400]}
df = pd.DataFrame(datos).set_index('Mes')

diagrama_ig(df)

plt.show()
```
